## ðŸ§  1. Conceptos Fundamentales de la Inteligencia Artificial (IA)

### Tipos de Aprendizaje en ML

1.  **Aprendizaje Supervisado:**
    * **DefiniciÃ³n:** El modelo se entrena con **datos etiquetados** (pares de entrada y salida conocidos).
    * **Ejemplo:** Predecir si un email es *spam* (clasificaciÃ³n) o el precio de una casa (regresiÃ³n).

2.  **Aprendizaje No Supervisado:**
    * **DefiniciÃ³n:** El modelo trabaja con **datos sin etiquetar** para encontrar patrones ocultos y estructuras internas.
    * **Ejemplo:** **Clustering** (agrupar clientes por comportamiento similar) o reducciÃ³n de dimensionalidad.

3.  **Aprendizaje por Refuerzo:**
    * **DefiniciÃ³n:** Un agente aprende a tomar decisiones en un entorno para **maximizar una recompensa acumulada**, a travÃ©s de prueba y error.

## ðŸ› ï¸ 2. Herramientas Esenciales de Python para Datos

El flujo de trabajo de ML comienza con el manejo eficiente de datos. Para ello, usamos:

### A. NumPy (Numerical Python)
Proporciona el objeto **`ndarray`** (array n-dimensional), que permite realizar **operaciones vectorizadas** de forma extremadamente rÃ¡pida, siendo la base de todo cÃ¡lculo numÃ©rico en Python.

### B. Pandas
Es la librerÃ­a mÃ¡s importante para la **manipulaciÃ³n y el anÃ¡lisis de datos**. Introduce el objeto **`DataFrame`**, que actÃºa como una tabla de datos (similar a una hoja de cÃ¡lculo o una tabla SQL).

```python
# =======================================================
# PASO 1: Importar LibrerÃ­as
# =======================================================
import numpy as np
import pandas as pd

print("âœ… LibrerÃ­as NumPy y Pandas importadas correctamente.")
print("-" * 30)

# =======================================================
# PASO 2: Ejemplo PrÃ¡ctico de NumPy
# =======================================================

# Crear un array NumPy
vector_numpy = np.array([10.5, 20.5, 30.5, 40.5])
print(f"1. Array NumPy (vector): {vector_numpy}")
print(f"   Tipo de datos: {vector_numpy.dtype}")

# OperaciÃ³n vectorizada (mucho mÃ¡s rÃ¡pido que un bucle 'for' en Python nativo)
# Se aplica la suma a CADA elemento del array
resultado_suma = vector_numpy + 5.0
print(f"2. Resultado de sumar 5.0 a cada elemento: {resultado_suma}")

print("-" * 30)

# =======================================================
# PASO 3: Ejemplo PrÃ¡ctico de Pandas
# =======================================================

# Crear un diccionario de datos
datos_alumnos = {
    'Nombre': ['Alice', 'Bob', 'Charlie', 'David'],
    'Edad': [25, 30, 35, 22],
    'Ciudad': ['NY', 'LA', 'Miami', 'NY'],
    'Nota_Final': [95.0, 88.5, 79.0, 99.0]
}

# Crear el DataFrame de Pandas
df_bootcamp = pd.DataFrame(datos_alumnos)

print("3. DataFrame creado (df_bootcamp):\n")
# Usamos head() para mostrar las primeras 5 filas (Ãºtil en grandes datasets)
print(df_bootcamp.head())

print("-" * 30)

# 4. OperaciÃ³n de Filtrado (SelecciÃ³n de filas)
# Queremos ver solo los alumnos que tienen una nota superior a 90.
alumnos_excelentes = df_bootcamp[df_bootcamp['Nota_Final'] > 90]

print("4. Alumnos con Nota_Final > 90:\n")
print(alumnos_excelentes)
```
