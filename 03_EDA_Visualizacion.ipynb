# üìä M√≥dulo 3: Visualizaci√≥n Exploratoria de Datos (EDA)

La **Visualizaci√≥n Exploratoria de Datos (EDA)** es una fase esencial para comprender la estructura de los datos, identificar tendencias y probar hip√≥tesis iniciales, a menudo utilizando m√©todos gr√°ficos.

## Conceptos Clave de la Visualizaci√≥n

### 1. Librer√≠as de Visualizaci√≥n en Python
* **Matplotlib:** La librer√≠a **fundamental** de Python para gr√°ficos. Permite un control muy fino sobre cada elemento del gr√°fico (t√≠tulos, ejes, colores, etc.).
* **Seaborn:** Est√° construida sobre Matplotlib. Ofrece una interfaz de alto nivel para generar gr√°ficos estad√≠sticos **m√°s atractivos** y complejos con menos c√≥digo.

### 2. Tipos de Gr√°ficos Comunes en EDA

| Tipo de Gr√°fico | Prop√≥sito | Variables (Ejes) |
| :--- | :--- | :--- |
| **Histograma** | Mostrar la **distribuci√≥n** de una √∫nica variable num√©rica (su frecuencia). | 1 Num√©rica |
| **Gr√°fico de Barras** | Mostrar la **frecuencia** o la suma de una variable categ√≥rica. | 1 Categ√≥rica |
| **Diagrama de Caja (Box Plot)** | Identificar la **dispersi√≥n**, la **mediana** y los valores **at√≠picos** (outliers) en una variable num√©rica. | 1 Num√©rica, opcionalmente 1 Categ√≥rica |
| **Diagrama de Dispersi√≥n (Scatter Plot)**| Mostrar la **relaci√≥n** entre dos variables num√©ricas. Es clave para ver correlaciones. | 2 Num√©ricas |
| **Mapa de Calor (Heatmap)** | Visualizar la **matriz de correlaci√≥n** entre todas las variables num√©ricas del DataFrame. | N Num√©ricas |

```python
# =======================================================
# PASO 1: Importar librer√≠as de visualizaci√≥n
# =======================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # La base de la visualizaci√≥n
import seaborn as sns           # Herramienta de alto nivel para gr√°ficos estad√≠sticos

# Configuraci√≥n b√°sica para que los gr√°ficos aparezcan en el notebook
%matplotlib inline
# Establece un estilo predeterminado para Seaborn para gr√°ficos m√°s limpios
sns.set_style("whitegrid")

print("‚úÖ Librer√≠as de visualizaci√≥n (Matplotlib y Seaborn) importadas.")
print("-" * 70)


# =======================================================
# PASO 2: Recrear y Limpiar el DataFrame (para asegurar la continuidad)
# =======================================================

# Usamos los datos del M√≥dulo 2
datos = {
    'Producto': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Precio_Unitario': [15.5, np.nan, 15.5, 22.0, 10.0, np.nan],
    'Cantidad': [3, 1, 3, 2, 4, 1],
    'Region': ['Norte', 'Sur', 'Norte', 'Oeste', 'Este', 'Sur']
}
df_ventas = pd.DataFrame(datos)

# Imputaci√≥n: Rellenamos los NaN con la mediana (Mejor pr√°ctica aprendida)
mediana_precio = df_ventas['Precio_Unitario'].median()
df_ventas['Precio_Unitario'] = df_ventas['Precio_Unitario'].fillna(mediana_precio)

print(f"1. DataFrame listo para visualizaci√≥n (NaN imputados con: {mediana_precio:.2f})")
print("-" * 70)


# =======================================================
# PASO 3: Visualizaci√≥n de Distribuci√≥n (Histograma y Box Plot)
# =======================================================

# Queremos ver c√≥mo se distribuyen los precios.

plt.figure(figsize=(10, 4)) # Define el tama√±o de la figura

# 3a. Histograma: Muestra la frecuencia de los precios
plt.subplot(1, 2, 1) # Define el espacio del primer gr√°fico (1 fila, 2 columnas, posici√≥n 1)
sns.histplot(df_ventas['Precio_Unitario'], kde=True, bins=5)
plt.title('Distribuci√≥n de Precios Unitarios (Histograma)')
plt.xlabel('Precio Unitario (‚Ç¨)')
plt.ylabel('Frecuencia')

# 3b. Diagrama de Caja (Box Plot): Identifica la dispersi√≥n y los at√≠picos
plt.subplot(1, 2, 2) # Define el espacio del segundo gr√°fico (1 fila, 2 columnas, posici√≥n 2)
sns.boxplot(y=df_ventas['Precio_Unitario'])
plt.title('Dispersi√≥n de Precios (Box Plot)')

plt.tight_layout() # Ajusta autom√°ticamente los par√°metros de la subtrama
plt.show() # Muestra los gr√°ficos
print("2. Distribuci√≥n visualizada con Histograma y Box Plot.")
print("-" * 70)


# =======================================================
# PASO 4: Visualizaci√≥n de la Relaci√≥n entre Variables
# =======================================================

# Queremos ver si hay una relaci√≥n entre el precio y la cantidad.

plt.figure(figsize=(7, 5))
# Diagrama de Dispersi√≥n (Scatter Plot)
sns.scatterplot(x='Precio_Unitario', y='Cantidad', data=df_ventas, hue='Region', s=100)
plt.title('Relaci√≥n entre Precio y Cantidad, por Regi√≥n')
plt.xlabel('Precio Unitario (‚Ç¨)')
plt.ylabel('Cantidad Vendida')
plt.legend(title='Regi√≥n')
plt.show() # Muestra el gr√°fico
print("3. Relaci√≥n visualizada con Scatter Plot, diferenciando por Regi√≥n.")
print("-" * 70)
```
